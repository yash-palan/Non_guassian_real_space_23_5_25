```python
import numpy as np
import matplotlib.pyplot as plt
fun = lambda omega,val: np.sqrt(omega/val)


vec_fun = np.vectorize(fun, otypes=[np.float64])
omega_vals = np.linspace(0,1,1001)
y = vec_fun(omega_vals, 2)
plt.plot(omega_vals, y, label='y = sqrt(omega * 2)')
plt.plot(omega_vals,  np.floor(10*y)/10, label='y = sqrt(omega * 2) rounded to 1 decimal place', linestyle='--')
plt.xlabel('omega')
plt.ylabel('y')
plt.title(r'Plot of \gamma_0 = sqrt(omega_0 * 2)')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fun = lambda gamma,val: gamma**2*val


vec_fun = np.vectorize(fun, otypes=[np.float64])
gamma_vals = np.linspace(0,1,10001)
omega = vec_fun(gamma_vals, 2)
plt.plot(gamma_vals, omega, label='omega')
plt.plot(gamma_vals, gamma_vals, linestyle='--',label = 'gamma')
plt.xlabel('gamma')
plt.ylabel('omega')
plt.legend()
# plt.title(r'Plot of \gamma_0 = sqrt(omega_0 * 0.4)')
# plt.legend()
```
